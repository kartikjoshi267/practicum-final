```python

import cv2
import time
cascade_src = 'cars.xml'
video_src = 'video3.MP4'
#line a
ax1=70
ay=90
ax2=230
#line b
bx1=15
by=125
bx2=225
#car num
i = 1
start_time = time.time()
#video ....
cap = cv2.VideoCapture(video_src)
car_cascade = cv2.CascadeClassifier(cascade_src)
def Speed_Cal(time):
    
    try:
        Speed = (2*9.144*3600)/(time*1000)
        return Speed
    except ZeroDivisionError:
        pass
while True:
    ret, img = cap.read()
    if (type(img) == type(None)):
        break
    #blurring to have exacter detection
    img = cv2.resize(img, (600,500))
    
    blurred = cv2.blur(img,ksize=(15,15))
    gray = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)
    cars = car_cascade.detectMultiScale(gray, 1.1, 2)
    
    #line a #i know road has got 
    cv2.line(img,(0,ay-50),(550,ay-50),(255,0,0),2)
    #line b
    cv2.line(img,(0,by-50),(550,by-50),(0,0,),2)
    
    for (x,y,w,h) in cars:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)   
        cv2.circle(img,(int((x+x+w)/2),int((y+y+h)/2)),1,(0,255,180),-1)
        
        while abs(int(ay) - int((y+y+h)/2)) < 2:
            start_time = time.time()
            break
            
        while int(ay) <= int((y+y+h)/2):
            if int(by) <= int((y+y+h)/2)&int(by+10) >= int((y+y+h)/2):
                # cv2.line(img,(bx1,by),(bx2,by),(0,255,0),2)
                Speed = int(Speed_Cal(time.time() - start_time))
                print("Car Number "+str(i)+" Speed: "+str(Speed)+" KM/H")
                i = i + 1
                cv2.putText(img, "Speed: "+str(Speed)+" KM/H", (x,y-15), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,210,80),3)
                break
            else :
                # cv2.putText(img, "Calculating", (100,200), cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,255,0),3)
                break
                
                
    cv2.imshow('video', img)
    
    if cv2.waitKey(33) & 0xff == ord('q'):
        break

cv2.destroyAllWindows()
```
